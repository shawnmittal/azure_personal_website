```python
import nltk
import re
import nltk
import spacy
from spacy import displacy
from pathlib import Path
from newspaper import Article
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.data.path.append(Path.cwd().parent / "web_app/nlp_example/nltk_data")
nlp = spacy.load('en_core_web_sm')
```

```python
def get_and_parse(url):
    article = Article(url)
    article.download()
    article.parse()
    article.nlp()
    return article
```

```python
def vader(article):
    sid = SentimentIntensityAnalyzer()
    return sid.polarity_scores(article.text)
```

```python
test_url = "https://www.npr.org/sections/goatsandsoda/2020/12/23/948157452/we-all-know-smoke-is-bad-for-your-health-it-could-be-worse-than-you-think"
```

```python
article = get_and_parse(test_url)
vader_score = vader(article)
```
